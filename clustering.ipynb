```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.decomposition import NMF
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from typing import Tuple, List, Dict

def create_user_embedding_matrix_optimized(user_items_df, batch_size=1000):
    # Select relevant columns
    df = user_items_df[['pid', 'corrected_food', 'embedding', 'time']]
    
    # Get unique users
    users = df['pid'].unique().tolist()
    print(f"Processing {len(users)} unique users")
    
    # Process embeddings
    if isinstance(df['embedding'].iloc[0], str):
        df['embedding_vector'] = df['embedding'].apply(lambda x: np.array(eval(x)))
    else:
        df['embedding_vector'] = df['embedding']
    
    # Get embedding dimension
    embedding_dim = len(df['embedding_vector'].iloc[0])
    
    # Create time bins
    df['time_bin'] = pd.cut(df['time'], bins=[0, 6, 12, 18, 24],
                           labels=['night', 'morning', 'afternoon', 'evening'])
    
    # Initialize matrix
    user_embedding_matrix = np.zeros((len(users), embedding_dim * 4))
    user_map = {user: i for i, user in enumerate(users)}
    
    # Process in batches
    for i in range(0, len(users), batch_size):
        if i % batch_size == 0:
            print(f"Processing users {i} to {min(i+batch_size, len(users))}")
        
        batch_users = users[i:i+batch_size]
        batch_df = df[df['pid'].isin(batch_users)]
        
        # Group by user and time bin
        for user in batch_users:
            user_idx = user_map[user]
            user_data = batch_df[batch_df['pid'] == user]
            
            for j, time_bin in enumerate(['night', 'morning', 'afternoon', 'evening']):
                bin_data = user_data[user_data['time_bin'] == time_bin]
                
                if len(bin_data) > 0:
                    bin_embeddings = np.vstack(bin_data['embedding_vector'].values)
                    avg_embedding = bin_embeddings.mean(axis=0)
                    
                    start_idx = j * embedding_dim
                    end_idx = (j + 1) * embedding_dim
                    user_embedding_matrix[user_idx, start_idx:end_idx] = avg_embedding
    
    return user_embedding_matrix, users


user_embedding_matrix, users = create_user_embedding_matrix_optimized(user_items)

```

```python

def extract_temporal_features(user_items_df: pd.DataFrame) -> Tuple[pd.DataFrame, List[str]]:
    """
    Extract temporal eating pattern features from food log data.
    
    Args:
        user_items_df: DataFrame containing user food log data with time
        
    Returns:
        temporal_features: DataFrame with temporal features per user
        users: List of user IDs
    """
    # Select relevant columns
    df = user_items_df[['pid', 'time']]
    
    # Create time bins
    time_bins = [0, 6, 12, 18, 24]
    bin_labels = ['night', 'morning', 'afternoon', 'evening']
    df['time_bin'] = pd.cut(df['time'], bins=time_bins, labels=bin_labels)
    
    # Get unique users
    users = df['pid'].unique().tolist()
    
    # Initialize features DataFrame
    temporal_features = pd.DataFrame({'pid': users})
    
    # Calculate percentage of meals in each time bin
    for bin_label in bin_labels:
        bin_counts = df[df['time_bin'] == bin_label].groupby('pid').size()
        bin_counts = bin_counts.reindex(users).fillna(0)
        
        total_counts = df.groupby('pid').size()
        total_counts = total_counts.reindex(users).fillna(1)  # Avoid division by zero
        
        percentage = (bin_counts / total_counts * 100).values
        temporal_features[f'pct_{bin_label}_meals'] = percentage
    
    # Calculate variance in eating times
    time_variance = df.groupby('pid')['time'].var().reindex(users).fillna(0)
    temporal_features['time_variance'] = time_variance.values
    
    # Calculate average time between meals per day
    # First, sort by user and time
    df_sorted = df.sort_values(['pid', 'time'])
    
    # Calculate time differences within each user
    df_sorted['time_diff'] = df_sorted.groupby('pid')['time'].diff()
    
    # Calculate average time difference
    avg_time_diff = df_sorted.groupby('pid')['time_diff'].mean().reindex(users).fillna(0)
    temporal_features['avg_time_between_meals'] = avg_time_diff.values
    
    return temporal_features, users


user_temporal_matrix, users = extract_temporal_features(user_items)
```

```python


def prepare_sleep_features(user_sleep_df: pd.DataFrame) -> Tuple[pd.DataFrame, List[str]]:
    """
    Aggregate sleep data into user features.
    
    Args:
        user_sleep_df: DataFrame containing user sleep data
        
    Returns:
        sleep_features: DataFrame with aggregated sleep features per user
        users: List of user IDs
    """
    # Select relevant columns
    df = user_sleep_df[['pid', 'sleep_time_decimal', 'sleep_duration_decimal', 'wakeup_time_decimal']]
    
    # Group by user and aggregate
    sleep_features = df.groupby('pid').agg({
        'sleep_time_decimal': ['mean', 'std'],
        'sleep_duration_decimal': ['mean', 'std'],
        'wakeup_time_decimal': ['mean', 'std']
    }).reset_index()
    
    # Flatten column names
    sleep_features.columns = ['pid'] + [
        f'{col[0]}_{col[1]}' for col in sleep_features.columns[1:]
    ]
    
    # Handle missing values (users with only one sleep record won't have std)
    sleep_features = sleep_features.fillna(0)
    
    users = sleep_features['pid'].tolist()
    
    return sleep_features, users

def apply_matrix_factorization(matrix: csr_matrix, n_components: int = 10) -> np.ndarray:
    """
    Apply Non-Negative Matrix Factorization to the user-food matrix.
    
    Args:
        matrix: Sparse user-food matrix
        n_components: Number of latent factors
        
    Returns:
        user_factors: Matrix of user representations in latent space
    """
    # Initialize NMF model
    model = NMF(
        n_components=n_components,
        init='random',
        random_state=42,
        max_iter=200
    )
    
    # Fit the model and return user factors
    user_factors = model.fit_transform(matrix)
    
    return user_factors

def combine_features(user_factors: np.ndarray, 
                    sleep_features: pd.DataFrame,
                    users_mf: List[str],
                    users_sleep: List[str]) -> pd.DataFrame:
    """
    Combine user factors from matrix factorization with sleep features.
    
    Args:
        user_factors: Matrix of user representations from MF
        sleep_features: DataFrame with sleep features
        users_mf: List of user IDs from matrix factorization
        users_sleep: List of user IDs from sleep features
        
    Returns:
        combined_df: DataFrame with combined features
    """
    # Create DataFrame from user factors
    user_factors_df = pd.DataFrame(
        user_factors, 
        columns=[f'factor_{i}' for i in range(user_factors.shape[1])]
    )
    user_factors_df['pid'] = users_mf
    
    # Merge with sleep features
    # Note: This will only include users who have both food and sleep data
    combined_df = user_factors_df.merge(sleep_features, on='pid', how='inner')
    
    return combined_df

def cluster_users(combined_df: pd.DataFrame, 
                 n_clusters: int = 5,
                 random_state: int = 42) -> Tuple[np.ndarray, float]:
    """
    Cluster users based on combined features.
    
    Args:
        combined_df: DataFrame with combined features
        n_clusters: Number of clusters
        random_state: Random seed
        
    Returns:
        labels: Cluster assignments
        score: Silhouette score
    """
    # Extract features (exclude pid)
    features = combined_df.drop('pid', axis=1).values
    
    # Standardize features
    scaler = StandardScaler()
    features_scaled = scaler.fit_transform(features)
    
    # Apply K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=random_state)
    labels = kmeans.fit_predict(features_scaled)
    
    # Evaluate clustering
    score = silhouette_score(features_scaled, labels)
    
    return labels, score

def find_optimal_clusters(combined_df: pd.DataFrame, 
                         max_clusters: int = 10) -> Dict:
    """
    Find optimal number of clusters using silhouette score.
    
    Args:
        combined_df: DataFrame with combined features
        max_clusters: Maximum number of clusters to try
        
    Returns:
        results: Dictionary with cluster counts and scores
    """
    results = {}
    for n_clusters in range(2, max_clusters + 1):
        labels, score = cluster_users(combined_df, n_clusters=n_clusters)
        results[n_clusters] = score
        print(f"Clusters: {n_clusters}, Silhouette Score: {score:.4f}")
    
    return results

def run_enhanced_pipeline(user_items_df: pd.DataFrame, 
                      user_sleep_df: pd.DataFrame,
                      n_clusters: int = 5,
                      batch_size: int = 1000) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Run the complete enhanced pipeline from data to clusters using embeddings and temporal features.
    
    Args:
        user_items_df: DataFrame with food log data including embeddings
        user_sleep_df: DataFrame with sleep data
        n_clusters: Number of user clusters
        batch_size: Batch size for processing user embeddings
        
    Returns:
        result_df: DataFrame with user IDs and cluster assignments
        combined_df: DataFrame with combined features
    """
    # Step 1: Create user embedding matrix
    print("Creating user embedding matrix from food embeddings...")
    user_embedding_matrix, users_emb = create_user_embedding_matrix_optimized(user_items_df, batch_size=batch_size)
    print(f"User embedding matrix shape: {user_embedding_matrix.shape}")
    
    # Step 2: Extract temporal eating pattern features
    print("Extracting temporal eating patterns...")
    temporal_features, users_temp = extract_temporal_features(user_items_df)
    print(f"Temporal features: {temporal_features.shape[0]} users, {temporal_features.shape[1] - 1} features")
    
    # Step 3: Prepare sleep features
    print("Preparing sleep features...")
    sleep_features, users_sleep = prepare_sleep_features(user_sleep_df)
    print(f"Sleep features: {sleep_features.shape[0]} users, {sleep_features.shape[1] - 1} features")
    
    # Step 4: Combine all features
    # First convert embedding matrix to DataFrame
    embedding_cols = [f'emb_{i}' for i in range(user_embedding_matrix.shape[1])]
    user_embedding_df = pd.DataFrame(
        user_embedding_matrix, 
        columns=embedding_cols
    )
    user_embedding_df['pid'] = users_emb
    
    # Merge with temporal features
    print("Combining features...")
    combined_df = user_embedding_df.merge(temporal_features, on='pid', how='inner')
    
    # Merge with sleep features
    combined_df = combined_df.merge(sleep_features, on='pid', how='inner')
    
    print(f"Combined features: {combined_df.shape[0]} users, {combined_df.shape[1] - 1} features")
    
    # Step 5: Cluster users
    print(f"Clustering users into {n_clusters} clusters...")
    labels, score = cluster_users(combined_df, n_clusters=n_clusters)
    print(f"Clustering complete. Silhouette score: {score:.4f}")
    
    # Step 6: Create result DataFrame
    result_df = pd.DataFrame({
        'pid': combined_df['pid'],
        'cluster': labels
    })
    
    return result_df, combined_df

# Example usage
result_df, combined_df = run_enhanced_pipeline(
    user_items_df=user_items, 
    user_sleep_df=user_sleep, 
    n_clusters=5,
    batch_size=1000
)
```

```python

# Additional analysis
def analyze_clusters(result_df: pd.DataFrame, 
                    combined_df: pd.DataFrame,
                    user_items_df: pd.DataFrame) -> Dict:
    """
    Analyze clusters to understand user patterns.
    
    Args:
        result_df: DataFrame with cluster assignments
        combined_df: DataFrame with combined features
        user_items_df: Original food log data
        
    Returns:
        cluster_profiles: Dictionary with cluster descriptions
    """
    # Add cluster assignments to combined features
    analysis_df = combined_df.copy()
    analysis_df['cluster'] = result_df['cluster']
    
    # Get cluster means for each feature
    cluster_means = analysis_df.groupby('cluster').mean()
    
    # Get food preferences per cluster
    merged_df = user_items_df.merge(result_df, on='pid')
    food_counts = merged_df.groupby(['cluster', 'corrected_food']).size().reset_index(name='count')
    
    # For each cluster, get top foods
    top_foods_by_cluster = {}
    for cluster in food_counts['cluster'].unique():
        cluster_foods = food_counts[food_counts['cluster'] == cluster]
        top_foods = cluster_foods.sort_values('count', ascending=False).head(10)
        top_foods_by_cluster[cluster] = top_foods
    
    # Create profiles
    cluster_profiles = {}
    for cluster in cluster_means.index:
        profile = {
            'sleep_time_mean': cluster_means.loc[cluster, 'sleep_time_decimal_mean'],
            'sleep_duration_mean': cluster_means.loc[cluster, 'sleep_duration_decimal_mean'],
            'wakeup_time_mean': cluster_means.loc[cluster, 'wakeup_time_decimal_mean'],
            'top_foods': top_foods_by_cluster[cluster]['corrected_food'].tolist()
        }
        cluster_profiles[cluster] = profile
    
    return cluster_profiles
```
